```python
def visualize_evolution_timeline():
    """æ·±å±¤å­¦ç¿’ã®é€²åŒ–ã®ã‚¿ã‚¤ãƒ ãƒ©ã‚¤ãƒ³"""
    fig, ax = plt.subplots(figsize=(18, 8))
    
    timeline = [
        (2012, 'AlexNet', 'CNNå¾©æ´»', 'lightblue'),
        (2014, 'VGGNet', 'æ·±ã„CNN', 'lightblue'),
        (2015, 'ResNet', 'ã‚¹ã‚­ãƒƒãƒ—æ¥ç¶š', 'lightgreen'),
        (2015, 'U-Net', 'ã‚»ã‚°ãƒ¡ãƒ³ãƒ†ãƒ¼ã‚·ãƒ§ãƒ³', 'lightgreen'),
        (2017, 'Transformer', 'è‡ªç„¶è¨€èªå‡¦ç†', 'lightyellow'),
        (2020, 'ViT', 'ç”»åƒã«Transformer', 'lightyellow'),
        (2020, 'NeRF', '3Dè¡¨ç¾', 'lightcoral'),
        (2023, '3DGS', 'é«˜é€Ÿ3D', 'lightcoral'),
    ]
    
    for i, (year, name, desc, color) in enumerate(timeline):
        x = (year - 2012) / 12
        y = 0.5 + 0.2 * ((-1) ** i)
        
        rect = plt.Rectangle((x, y - 0.08), 0.08, 0.16, 
                            facecolor=color, edgecolor='gray', linewidth=2)
        ax.add_patch(rect)
        ax.text(x + 0.04, y + 0.12, str(year), ha='center', fontsize=9)
        ax.text(x + 0.04, y, name, ha='center', va='center', fontsize=10, fontweight='bold')
        ax.text(x + 0.04, y - 0.12, desc, ha='center', fontsize=8)
    
    # æ™‚é–“è»¸
    ax.axhline(y=0.5, color='black', linewidth=1, linestyle='-')
    
    # å‡¡ä¾‹
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='lightblue', edgecolor='gray', label='CNNæ™‚ä»£'),
        Patch(facecolor='lightgreen', edgecolor='gray', label='ã‚¢ãƒ¼ã‚­ãƒ†ã‚¯ãƒãƒ£é©æ–°'),
        Patch(facecolor='lightyellow', edgecolor='gray', label='Transformeræ™‚ä»£'),
        Patch(facecolor='lightcoral', edgecolor='gray', label='3Dç†è§£'),
    ]
    ax.legend(handles=legend_elements, loc='upper left', fontsize=10)
    
    ax.set_xlim(-0.05, 1)
    ax.set_ylim(0.2, 0.85)
    ax.axis('off')
    ax.set_title('æ·±å±¤å­¦ç¿’ã«ã‚ˆã‚‹è¦–è¦šç†è§£ã®é€²åŒ–', fontsize=16, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

visualize_evolution_timeline()
```
